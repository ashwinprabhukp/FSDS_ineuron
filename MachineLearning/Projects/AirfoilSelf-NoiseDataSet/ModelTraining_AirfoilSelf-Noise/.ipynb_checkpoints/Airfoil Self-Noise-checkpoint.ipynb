## https://archive.ics.uci.edu/ml/datasets/Airfoil+Self-Noise

```python
import pandas as pd
```

```python
df = pd.read_csv('airfoil_self_noise.dat', sep='\t', header=None)
```

```python
# Define the columns
df.columns = ["Frequency", "Angle of attack","Chord length", "Free-stream velocity", "Suction side", "pressure level"]
```

```python
df.head()
```

```python
# Check if there are null in data
df.isnull().sum()
```

```python
# Train test split
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

## Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV

regressor = LinearRegression()

# initialize parameters
paramenters= dict()
paramenters['fit_intercept'] = [False, True]
# paramenters['normalize'] = [False, True] ---> Deprecated 

linear_regressor = GridSearchCV(regressor, paramenters, scoring='neg_mean_squared_error', cv=5,verbose=1)

#Fit the model
linear_regressor.fit(X_train, y_train)

# Calculation of R2 Score
from sklearn.metrics import r2_score
y_pred = linear_regressor.predict(X_test)
r2_score = r2_score(y_test, y_pred)
print('R2 score: ' + str(r2_score))

# calculate the best params and score
print(linear_regressor.best_params_)
print(linear_regressor.best_score_)
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

ridge = Ridge()

# initialize parameters
parameters= dict()
parameters['alpha'] = [1e-15,1e-3,1e-2,1,5,10,20,30,40,50,55,100]
parameters['solver'] = ['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga']
parameters['fit_intercept'] = [False, True]

ridge_regressor = GridSearchCV(ridge, parameters, scoring='neg_mean_squared_error', cv=5,verbose=1)

#Fit the model
ridge_regressor.fit(X_train, y_train)

# Calculation of R2 Score
from sklearn.metrics import r2_score
y_pred = ridge_regressor.predict(X_test)
r2_score = r2_score(y_test, y_pred)
print('R2 score: ' + str(r2_score))

# calculate the best params and score
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

## Lasso Regression

```python
from numpy import random
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

lasso=Lasso()

# initialize parameters
parameters= dict()
parameters['alpha'] = [1e-5,1e-3,1e-2,1,5,10,20,30,40,50,55]
parameters['selection'] = ['cyclic', 'random']
parameters['fit_intercept'] = [False, True]
parameters['warm_start'] = [False, True]
parameters['precompute'] = [False, True]
parameters['max_iter'] = random.randint(low=100, high=10000, size=10)

lasso_regressor=GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)

#Fit the model
lasso_regressor.fit(X_train, y_train)

# Calculation of R2 Score
from sklearn.metrics import r2_score
y_pred = lasso_regressor.predict(X_test)
r2_score = r2_score(y_test, y_pred)
print('R2 score: ' + str(r2_score))

# calculate the best params and score
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

## Elastic regression

```python
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV

elasticnet=ElasticNet()

# initialize parameters
parameters= dict()
parameters['alpha'] = [1e-9, 1e-8,1e-5,1e-2,1,5,10,20,30,35,40,45,50,55,100]
parameters['selection'] = ['cyclic', 'random']
parameters['fit_intercept'] = [False, True]
parameters['warm_start'] = [False, True]
parameters['precompute'] = [False, True]
#parameters['max_iter'] = random.randint(low=100, high=10000, size=5)
parameters['l1_ratio'] = random.uniform(low=0.0, high=1.0, size=5)

elasticnet_regressor=GridSearchCV(elasticnet,parameters,scoring='neg_mean_squared_error',cv=5)


#Fit the model
elasticnet_regressor.fit(X_train, y_train)

# Calculation of R2 Score
from sklearn.metrics import r2_score
y_pred = elasticnet_regressor.predict(X_test)
r2_score = r2_score(y_test, y_pred)
print('R2 score: ' + str(r2_score))

# calculate the best params and score
print(elasticnet_regressor.best_params_)
print(elasticnet_regressor.best_score_)
```

## decission tree regressor

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV

Regressor=DecisionTreeRegressor()

# initialize parameters
parameters= dict()
parameters['ccp_alpha'] = [1e-25,1e-20,1e-15,1e-10,1e-5,1,5,10]
parameters['splitter'] = ['best', 'random']
parameters['criterion'] = ['squared_error', 'friedman_mse', 'absolute_error', 'poisson']
parameters['max_features'] = ['auto', 'sqrt', 'log2']
parameters['max_depth'] = [5,10,15,20]

decisiontree_regressor=GridSearchCV(Regressor,parameters,scoring='neg_mean_squared_error',cv=5)


#Fit the model
decisiontree_regressor.fit(X_train, y_train)

# Calculation of R2 Score
from sklearn.metrics import r2_score
y_pred = decisiontree_regressor.predict(X_test)
r2_score = r2_score(y_test, y_pred)
print('R2 score: ' + str('%.3f'%r2_score))

# calculate Adjusted R2 Score
p = len(X_train.columns) # number of features
n = len(X_train) # number of observations
Adj_r2 = 1-(1-r2_score)*(n-1)/(n-p-1)
print('Adjusted R2 score: ' + str('%.3f'%Adj_r2))

# calculate the best params and score
print(decisiontree_regressor.best_params_)
print(decisiontree_regressor.best_score_)
```

## Random forest regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

Regressor=RandomForestRegressor()

# initialize parameters
parameters= dict()
# parameters['n_estimators'] = [10,20,80,100,200,500]
# parameters['criterion'] = ['squared_error', 'friedman_mse', 'absolute_error', 'poisson']
# parameters['max_features'] = ['auto', 'sqrt', 'log2']
# parameters['max_depth'] = [5,10,15,20]
# parameters['min_samples_split'] = [1,2,3,4,5]

randomforest_regressor=GridSearchCV(Regressor,parameters,scoring='neg_mean_squared_error',cv=5)


#Fit the model
randomforest_regressor.fit(X_train, y_train)

# Calculation of R2 Score
from sklearn.metrics import r2_score
y_pred = randomforest_regressor.predict(X_test)
r2_score = r2_score(y_test, y_pred)
print('R2 score: ' + str(r2_score))

# calculate the best params and score
print(randomforest_regressor.best_params_)
print(randomforest_regressor.best_score_)
```

## Support Vector Regressor (SVR)

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV

Regressor=SVR()

# initialize parameters
parameters= dict()
# parameters['degree'] = [1,2,3,4,5]
# parameters['kernel'] = ['linear', 'poly', 'rbf', 'sigmoid', 'precomputed']
# parameters['gamma'] = ['scale', 'auto']

svr_regressor=GridSearchCV(Regressor,parameters,scoring='neg_mean_squared_error',cv=5)


#Fit the model
svr_regressor.fit(X_train, y_train)

# Calculation of R2 Score
from sklearn.metrics import r2_score
y_pred = svr_regressor.predict(X_test)
r2_score = r2_score(y_test, y_pred)
print('R2 score: ' + str(r2_score))

# calculate Adjusted R2 Score
p = len(X_train.columns) # number of features
n = len(X_train) # number of observations
Adj_r2 = 1-(1-r2_score)*(n-1)/(n-p-1)
print('Adjusted R2 score: ' + str('%.3f'%Adj_r2))

# calculate the best params and score
print(svr_regressor.best_params_)
print(svr_regressor.best_score_)
```
